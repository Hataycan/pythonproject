```python
import requests
from bs4 import BeautifulSoup
import yfinance as yf

# Scrape Tesla's revenue data
url_revenue = "https://companiesmarketcap.com/tesla/revenue/"
response_revenue = requests.get(url_revenue)
soup_revenue = BeautifulSoup(response_revenue.content, 'html.parser')
table_revenue = soup_revenue.find('table', class_='table')
rows_revenue = table_revenue.find_all('tr')
years = []
revenues = []

for row in rows_revenue:
    columns = row.find_all('td')
    if len(columns) >= 2:
        year = columns[0].text.strip()
        revenue = columns[1].text.strip()
        years.append(year)
        revenues.append(revenue)

revenues = [float(revenue.replace('$', '').replace('B', '')) for revenue in revenues]
reversed_years = years[::-1]
reversed_revenues = revenues[::-1]

# Create a DataFrame for Tesla's revenue data
import pandas as pd
tesla_revenue = pd.DataFrame({'Year': reversed_years, 'Revenue': reversed_revenues})

# Display the last few rows of the tesla_revenue DataFrame
print(tesla_revenue.tail())

# Get Tesla's stock data using yfinance
tesla_symbol = 'TSLA'
tesla = yf.Ticker(tesla_symbol)
tesla_data = tesla.history(period="max")

# Display the first few rows of the tesla_data DataFrame
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://companiesmarketcap.com/gamestop/revenue/"

# Send a GET request to the URL and get the HTML content
response = requests.get(url)
html_content = response.content

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table containing the revenue data
table = soup.find('table', class_='table')

# Find all rows within the tbody
rows = table.find_all('tr')

# Initialize lists to store the extracted data
years = []
revenues = []

# Extract data from the rows
for row in rows:
    columns = row.find_all('td')
    if len(columns) >= 2:
        year = columns[0].text.strip()
        revenue = columns[1].text.strip()
        years.append(year)
        revenues.append(revenue)

# Convert revenue values to numeric
revenues = [float(revenue.replace('$', '').replace('B', '')) for revenue in revenues]

# Create a DataFrame for GameStop's revenue data
gme_revenue = pd.DataFrame({'Year': years, 'Revenue': revenues})

# Display the last few rows of the gme_revenue DataFrame using the tail function
print(gme_revenue.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the ticker symbol for Tesla
tesla_symbol = 'TSLA'

# Create a Ticker object for Tesla
tesla_ticker = yf.Ticker(tesla_symbol)

# Get historical stock data for Tesla
tesla_data = tesla_ticker.history(period="1y")

# Function to make a graph
def make_graph(data, ylabel, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'], marker='o')
    plt.xlabel('Date')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Use the make_graph function to plot Tesla's stock price data
make_graph(tesla_data, 'Stock Price ($)', 'Tesla Stock Price Over Time')
```
